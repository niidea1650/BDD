```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Ruta del archivo CSV
ruta_archivo = '/content/bball.csv'
csv = pd.read_csv(ruta_archivo, sep=',')
```

```python
# Cargar el archivo CSV en un DataFrame
df = pd.read_csv(ruta_archivo)
```

```python
# Contar jugadores retirados y activos
jugadores_retirados = df[df['Retired'] == 'Y']['Retired'].count()
jugadores_activos = df[df['Retired'] == 'N']['Retired'].count()
```

```python
print("Jugadores retirados:", jugadores_retirados)
print("Jugadores activos:", jugadores_activos)
```

```python
# Etiquetas y valores para el gráfico de torta
labels = ['Retirados', 'Activos']
sizes = [jugadores_retirados, jugadores_activos]
colors = ['#074173', '#1679AB']  # Colores más vibrantes
explode = (0, 0)  # Explode para resaltar la categoría de "Retirados"
```

```python
# Crear el gráfico de torta
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.f%%', shadow=True, startangle=90)
plt.title('jugadores retirados y activos', fontsize=16, fontweight='bold')  # Título más grande y en negrita

```

```python
top_jugadores = csv.sort_values(by='Total PTS', ascending=False).head(50)
```

```python
# Crear el gráfico de puntos
plt.figure(figsize=(10, 6))  # Tamaño del gráfico
plt.scatter(top_jugadores['Name'], top_jugadores['Total PTS'], color='blue')  # Graficar puntos
plt.xticks(rotation=90)  # Rotar etiquetas del eje x para mejor visualización
plt.xlabel('Jugador')  # Etiqueta eje x
plt.ylabel('Puntos Totales')  # Etiqueta eje y
plt.title('Puntos Totales de los 50 primeros Jugadores')  # Título del gráfico
plt.grid(True)  # Mostrar cuadrícula en el gráfico
```

```python
# Filtrar jugadores retirados y activos
jugadores_retirados = df[df['Retired'] == 'Y']
jugadores_activos = df[df['Retired'] == 'N']

# Obtener nombres y puntos totales de los jugadores
nombres_retirados = jugadores_retirados['Name']
puntos_retirados = jugadores_retirados['Total PTS']

nombres_activos = jugadores_activos['Name']
puntos_activos = jugadores_activos['Total PTS']
```

```python
# Obtener los 10 jugadores con más minutos jugados
top_jugadores = df.nlargest(10, 'MP')

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(top_jugadores['Name'], top_jugadores['MP'], color='skyblue')
plt.xlabel('Jugador')
plt.ylabel('Minutos Jugados')
plt.title('Top 10 Jugadores con Más Minutos Jugados y su Pick en el Draft')

# Agregar la información del pick en el draft como anotaciones en el gráfico
for i, mp in enumerate(top_jugadores['MP']):
    plt.text(i, mp + 0.5, f'Pick: {top_jugadores.iloc[i]["Pick"]}', ha='center', va='bottom', rotation=45)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Crear el gráfico de puntos (scatter plot)
plt.figure(figsize=(10, 6))
plt.scatter(top_jugadores['Pick'], top_jugadores['MP'], color='skyblue')
plt.xlabel('Pick en el Draft')
plt.ylabel('Minutos Jugados')
plt.title('Top 10 Jugadores con Más Minutos Jugados y su Pick en el Draft')

# Agregar etiquetas de los jugadores
for i, name in enumerate(top_jugadores['Name']):
    plt.text(top_jugadores.iloc[i]['Pick'], top_jugadores.iloc[i]['MP'], name, ha='right', va='bottom', rotation=45)

plt.xticks(top_jugadores['Pick'])
plt.tight_layout()
plt.show()
```
